## (c) vi. Further exploration

Lets examine the elite schools more closely.

```python
college['AcceptRate'] = college.Accept/college.Apps
```

```python
college.describe() # examine acceptance rates
```

```python
college.boxplot(column='AcceptRate',by='Elite', figsize=(7,7));
```

So the acceptance rate is about 25% lower than non-elite schools

Lets look at the Expenditures per year at elite vs. non-elite

```python
college.boxplot(column='Expend',by='Elite', figsize=(7,7));
```

So an average of about 5,000 more per year at an Elite.

We can also look at the percentage that enroll for elite and non-elites

```python
college['EnrollRate'] = college.Enroll/college.Accept
```

```python
college.boxplot(column='EnrollRate',by='Elite', figsize=(7,7));
```

This is interesting, even though the elite schools have a lower acceptance rate (and thus more exclusive) this does not translate into a higher enrollment rate. The added cost could be one of the reasons for this...
